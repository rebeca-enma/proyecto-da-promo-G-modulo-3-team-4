```python
# Cambiamos los datos de las columnas de string a integer: 
columnas_num = ['MONTHLYRATE', 'NUMCOMPANIESWORKED', 'PERCENTSALARYHIKE',
       'PERFORMANCERATING', 'RELATIONSHIPSATISFACTION', 'STANDARDHOURS',
       'STOCKOPTIONLEVEL']

# son numericos: MONTHLYRATE', 'NUMCOMPANIESWORKED', 'PERCENTSALARYHIKE', 'PERFORMANCERATING', 'RELATIONSHIPSATISFACTION', 'STANDARDHOURS','STOCKOPTIONLEVEL'
#son booleanos : 'OVERTIME'
```

```python
def cambiar_columnas(cadena):
    if isinstance(cadena, int) or isinstance(cadena, float):  # Verificar si la cadena es un entero o un float
        return cadena
    elif ',' in cadena:
        try:
            return int(cadena.split(',')[0]) 
        except:
            try:
                return float(cadena.split(',')[0])
            except:
                return np.nan
    else:
        try:
            return int(cadena)
        except:
            try:
                return float(cadena)
            except:
                return np.nan
```

```python
for col in columnas_num:
    
    df_data2[col] = df_data2[col].apply(cambiar_columnas)
```
