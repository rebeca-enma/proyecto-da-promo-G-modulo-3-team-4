# Comprobación por columnas

```python
# Reemplazamos los números escritos en letras e ingles por integers
def transformar_age(dataframe,columna): 

    for i, age in enumerate (dataframe[columna]): 
        
        try:
            age = int(age)
            dataframe.loc[i, columna] = age 

        except: 
            numero_entero = w2n.word_to_num(age)  
            dataframe.loc[i, columna] = numero_entero
       
```

```python
# Actualmente se encuentran los datos de forma correcta, presentamos la función futura para dicha columna
def transformar_atr(dataframe,columna): 
    
    for i, atr in enumerate (dataframe[columna]): 
        
        try:
            if atr.lower() == 'yes' or 'no':  
                dataframe.loc[i, columna] = atr
   
            else: 
                dataframe.loc[i, columna] = None
                  
        except: 
            print('Error de datos')
 
```

```python
def transformar_bus_travel (dataframe, columna):

    dataframe[columna] = dataframe[columna].str.replace("_", " ").replace("-", " ")
```

```python
def transformar_daily(dataframe,columna): 
    
    for i, element in enumerate (dataframe[columna]): 
        
        if element == "nan$" or element == "nan" or element == "NaN":
            element = np.nan
            dataframe.loc[i, columna] = element
        elif type(element) == str:
            try:
                element = element.split(",")[0]
                int(element)
                dataframe.loc[i, columna] = element
            except: 
                print(f"Error daily {i, element}")
        else:
            continue
    
```

```python
def transformar_department (dataframe, columna):

    dataframe[columna] = dataframe[columna].str.lower().str.strip()
```

```python
def transformar_gender (dataframe, columna):
    mapeo = {0 : "male", 1 : "female"}
    dataframe[columna]= dataframe[columna].map(mapeo)     
```

```python
def limpieza_df(dataframe):
    
    for columna in dataframe.columns:
       

        if columna == "AGE":
            transformar_age(dataframe, "AGE")
        elif columna == "BUSINESSTRAVEL":
            transformar_bus_travel(dataframe, "BUSINESSTRAVEL")  
        elif columna == "DAILYRATE" or columna == "EMPLOYEENUMBER" or columna == "MONTHLYINCOME":
            for col in ["DAILYRATE", "EMPLOYEENUMBER","MONTHLYINCOME" ]:
                transformar_daily(dataframe, col)
        elif columna == "DEPARTMENT" or columna == "EDUCATIONFIELD" or columna == "ATTRITION" or columna == "JOBROLE":
            for colu in ["ATTRITION", "DEPARTMENT", "EDUCATIONFIELD","JOBROLE" ]:
                transformar_department(dataframe, colu)
        elif columna == "GENDER":
            transformar_gender(dataframe, "GENDER")
        
        else:
            continue
```

```python
# Llamamos a la funcion de LIMPIEZA del DF
limpieza_df(df_data2)

# En las funciones estamos cambiando el DF original por lo tanto las funciones no tienen return y por eso aqui mostramos como queda el DF modificado
df_data2.head()
```
