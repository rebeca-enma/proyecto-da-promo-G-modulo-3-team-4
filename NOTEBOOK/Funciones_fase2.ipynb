```python
# Reemplazamos los números escritos en letras e ingles por integers
def transformar_age(df,columna): 

    for i, age in enumerate (df[columna]): 
        
        try:
            age = int(age)
            df.loc[i, columna] = age 

        except: 
            numero_entero = w2n.word_to_num(age)  
            df.loc[i, columna] = numero_entero
       
```

```python
x = transformar_age(df_data2, "AGE")
```

```python
# Actualmente se encuentran los datos de forma correcta, presentamos la función futura para dicha columna
def transformar_atr(df,columna): 
    
    for i, atr in enumerate (df[columna]): 
        
        try:
            if atr.lower() == 'yes' or 'no':  
                df.loc[i, columna] = atr
   
            else: 
                df.loc[i, columna] = None
                  
        except: 
            print('Error de datos')
 
```

```python
x = transformar_atr(df_data2, "ATTRITION")
```

```python
def transformar_daily(df,columna): 
    
    for i, element in enumerate (df[columna]): 
        
        if type(element) == "class 'str'":
            try:
                element = element.split(",")[0]
                int(element)
                df.loc[i, columna] = element
            except: 
                print(f"Error de datos {i, element}")
        
        else:
            continue
    
```
