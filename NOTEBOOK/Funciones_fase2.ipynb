# Comprobación por columnas

```python
# Reemplazamos los números escritos en letras e ingles por integers
def transformar_age(dataframe,columna): 

    for i, age in enumerate (dataframe[columna]): 
        
        try:
            age = int(age)
            dataframe.loc[i, columna] = age 

        except: 
            numero_entero = w2n.word_to_num(age)  
            dataframe.loc[i, columna] = numero_entero
       
```

```python
# Actualmente se encuentran los datos de forma correcta, presentamos la función futura para dicha columna
def transformar_atr(dataframe,columna): 
    
    for i, atr in enumerate (dataframe[columna]): 
        
        try:
            if atr.lower() == 'yes' or 'no':  
                dataframe.loc[i, columna] = atr
   
            else: 
                dataframe.loc[i, columna] = None
                  
        except: 
            print('Error de datos')
 
```

```python
def transformar_daily(dataframe,columna): 
    
    for i, element in enumerate (dataframe[columna]): 
        
        if type(element) == "class 'str'":
            try:
                element = element.split(",")[0]
                int(element)
                dataframe.loc[i, columna] = element
            except: 
                print(f"Error de datos {i, element}")
        
        else:
            continue
    
```

```python
def limpieza_df(dataframe):
    
    for columna in dataframe.columns:
       

        if columna == "AGE":
            transformar_age(dataframe, "AGE")
        elif columna == "ATTRITION":
            transformar_atr(dataframe, "ATTRITION")
        elif columna == "DAILYRATE":
            transformar_daily(dataframe, "DAILYRATE")
        
        else:
            continue
```

El parámetro errors='coerce' en la función pd.to_numeric() de pandas se utiliza para manejar los errores que pueden surgir al intentar convertir valores no numéricos a números. Cuando se establece errors='coerce', significa que si hay algún valor que no se puede convertir a un número, se establecerá como NaN (Not a Number) en lugar de lanzar un error. Esto permite que el proceso de conversión continúe sin interrupciones, y luego puedes manejar los valores NaN según sea necesario.

```python
# Columna 11 - HOURLYRATE
# Convertir los valores string a integer, remplazando los valores 'Not Available' por NaN.
df_data2['HOURLYRATE'] = pd.to_numeric(df_data2['HOURLYRATE'], errors='coerce')
```

```python
def transformar_hourly(df_data2, columna):
    for i, element in enumerate(df_data2[columna]):
        if isinstance(element, str):
            try:
                value = int(element.split(",")[0])
                df_data2.loc[i, columna] = value
            except ValueError:
                df_data2.loc[i, columna] = None
                print(f"Error de datos en fila {i}: {element}")
        else:
            continue

# Llamada a la función
ratio_hora = transformar_hourly(df_data2, 'HOURLYRATE')

# Verificación del resultado
print(df_data2['HOURLYRATE'].head())
```

```python
# columna 12 - JOBINVOLMENT
# Convertir los valores string a integer, siguiendo una escala numérica del 1 al 4.
df_data2['JOBINVOLMENT'] = pd.to_numeric(df_data2['JOBINVOLVEMENT'], errors='coerce')
```

```python
def transformar_job_involvement(dataframe, columna):
    for i, element in enumerate(dataframe[columna]):
        try:
            value = int(element)
            if value < 1 or value > 4:
                raise ValueError("El valor está fuera del rango permitido.")
            dataframe.loc[i, columna] = value
        except (ValueError, TypeError):
            dataframe.loc[i, columna] = None
            print(f"Error de datos en fila {i}: {element}")

# Llamada a la función
transformar_job_involvement(df_data2, 'JOBINVOLVEMENT')

# Verificación del resultado
print(df_data2['JOBINVOLVEMENT'].head())
```
