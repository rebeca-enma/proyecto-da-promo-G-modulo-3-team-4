# Comprobación por columnas

```python
# Comprobación de la columna Age
lista_age = df_data2["AGE"].to_list()
```

```python
lista_age
```

```python
# Reemplazamos los números escritos en letras e ingles por integers
def transformar_age(lista): 
    lista_anios = []
    
    for age in lista: 
        
        try:
            age = int(age)
            lista_anios.append(age)  
        except: 
            numero_entero = w2n.word_to_num(age)  
            lista_anios.append(numero_entero)  
    
    return lista_anios         
```

```python
anios = transformar_age(lista_age)
print(anios)
len(anios)
```

```python
# Comprobación de la columna Attrition
lista_atr = df_data2["ATTRITION"].to_list()
lista_atr
```

```python
# Actualmente se encuentran los datos de forma correcta, presentamos la función futura para dicha columna
def transformar_atr(lista): 
    lista_atrit = []
    
    for atr in lista: 
        
        try:
            if atr.lower() == 'yes' or 'no': 
                lista_atrit.append(atr)  
              
            else: 
                lista_atrit.append(None)   
        except: 
            print('Error de datos')
    
    return lista_atrit
```

```python
atrit = transformar_atr(lista_atr)
atrit
len(atrit)
```
