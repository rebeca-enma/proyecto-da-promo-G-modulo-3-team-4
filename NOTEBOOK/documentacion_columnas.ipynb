Se decide sustituir los nulos con una nueva categora: "Unknown". Ya que utilizar la moda (valor ms comn) podra sesgar la distribucin de las categoras y se podra distorsionar resultados.

### Gestin de nulos numricos

   - En la columna "WORKLIFEBALANCE" imputamos los valores nulos con la moda, esto nos ayuda a preservar la distribucin de los datos y evita sesgar los resultados del anlisis hacia valores atpicos.
