### Gestión de nulos categóricos

Se decide sustituir los nulos con una nueva categoría: "Unknown". Ya que utilizar la moda (valor más común) podría sesgar la distribución de las categorías y se podría distorsionar resultados.

Al asignar "Unknown" nos permite mantener la integridad de los datos sin introducir información falsa o sesgada y así nos aseguramos el evitar cualquier impacto en los resultados de las consultas a la base de datos.

### Gestión de nulos numéricos

   - En el caso de la columna "EMPLOYEENUMBER" vemos que existen valores nulos, lo que podría ser debido a la falta de información o a la naturaleza de los datos. En este caso, decidimos reemplazar los valores nulos con "Unknown" para mantener la integridad de los datos y evita introducir información incorrecta.

   - Para las columnas "DAILYRATE", "PERFORMANCERATING" y "TOTALWORKINGYEARS", decidimos imputar valores en lugar de eliminar las filas con valores nulos, ya que estos pueden contener información valiosa para el análisis. 
   
   Utilizamos el método de imputación IterativeImputer, con esto sustituimos los valores nulos por datos numéricos con valores estimados basados en el resto de la información disponible en el conjunto de datos.

   - En la columna "WORKLIFEBALANCE" imputamos los valores nulos con la moda, esto nos ayuda a preservar la distribución de los datos y evita sesgar los resultados del análisis hacia valores atípicos.

En resumen, estas decisiones están respaldadas por la necesidad de mantener la integridad de los datos, conservar la cantidad de datos disponibles para el análisis y evitar sesgar los resultados del análisis debido a la imputación de valores inapropiados. Siendo coherentes con el contexto de los datos y los objetivos del análisis, estas estrategias ayudan a garantizar que los resultados obtenidos sean válidos y significativos.


## Gestión de duplicados 

Durante el proceso de gestión de duplicados en nuestra base de datos, hemos identificado que existen registros donde se repiten valores en la columna "employee number", indicando la presencia de múltiples registros para un mismo empleado. Sin embargo, en estos casos, observamos que los valores en otras columnas varían, lo que sugiere que se han tomado múltiples registros del mismo empleado en diferentes momentos.

Para comprender mejor estos registros y garantizar la integridad de nuestra base de datos, hemos llevado a cabo una investigación detallada para identificar las columnas en las que se producen dichos cambios. Este análisis nos ha ayudado a elaborar una base de datos que refleje con precisión los registros realizados al mismo empleado en por lo menos dos ocasiones, lo que proporciona una visión más completa y precisa de la información contenida en nuestra base de datos.
