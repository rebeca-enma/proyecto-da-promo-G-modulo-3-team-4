```python
for columna in list(df_data.columns):

    print(f" \n----------- ESTAMOS ANALIZANDO LA COLUMNA: '{columna.upper()}' -----------\n")
    print(f"* N¬∫ de datos: {len(df_data[columna].to_list())}")
    print(f"* Frecuencia de valores en la columna: \n {df_data[columna].value_counts()}")
    print(f"* Datos unicos en la columna {len(df_data[columna].unique())}")
    print(f"* Los valores son de tipo: {type(columna)}")
    print(f"La suma de datos nulos {df_data[columna].isnull().sum()}")
    print(df_data[columna].unique()) 
```
